```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import sys
sys.path.append("../")
import utils as ut
```

```python
peak_dir = "/data5/deepro/starrseq/main_library/3_peak_call/data"
lib_names = ["CC", "ATF2", "CTCF", "FOXA1", "LEF1", "SCRT1", "TCF7L2", "16P12_1"]
```

```python
def get_file_length(filename):
    f = open(filename, "r")
    lenf = len(f.readlines())
    f.close()
    return lenf


def get_peak_count_df(lib_names):
    peak_files = [ut.get_peak_file(peak_dir, ln) for ln in lib_names]
    counts = [get_file_length(pf) for pf in peak_files]
    df = pd.DataFrame({"library": lib_names, "peaks": counts})
    return df
```

```python
peak_df = get_peak_count_df(lib_names)
```

```python

sns.set(font_scale = 2)
```

```python
sns.set_style("whitegrid")
ax = sns.catplot(data=peak_df, x="library", y="peaks", aspect=1.25, height=8, palette="Set3", kind="bar")
ax.set_xticklabels(rotation=45)
ax.savefig("/data5/deepro/starrseq/main_library/4_quality_control_peaks/data/figures/peak_num.svg")
```
